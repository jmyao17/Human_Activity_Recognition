# Load dataset

- **There are 15 participants**, corresponding to 15 csv files.

Here I define a function to load all the 15 datasets into 15 separate DataFrames systematically.

```python
def load_datasets(filename):
    df = pd.read_csv(filename)
    return df
```

```python
dfs=[]
features=['Sequential_number', 'x_acceleration', 'y_acceleration', 'z_acceleration', 'label']
for i in range(1,16):
    filename='Activity Recognition from Single Chest-Mounted Accelerometer/'+str(i)+'.csv'
    df = load_datasets(filename)
    df.columns = [features]
    print('Time duration for the {}-th participant is {} seconds'.format(i,np.ceil(len(df['label'])/52)))
    dfs.append(df)
```
