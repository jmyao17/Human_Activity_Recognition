# Segmentation

```python
# parameters for the segmentation of the whole data

num_feature = 3
window_size = 520  # N_time_steps
step_size   = 52*5
segments    = []
label       = []
```

```python
# for  each participant
def get_segmentations(df):
    
    segments = []
    labels = []
    
    # loop over class [1,2,..,7]
    for class_label in range(1,8):
        
        df_class = df[df['label'].values == class_label]
        
        assert len(df_class) > window_size
        
        
        # for each  given class, slide the window
        #for i in range(0,len(df_class)- window_size, step_size): 
        for i in range(0,len(df_class) - window_size, step_size):

            start = i
            end   = i + window_size

            xseg = df_class['x_acceleration'].values[start: end]
            yseg = df_class['y_acceleration'].values[start: end]
            zseg = df_class['z_acceleration'].values[start: end]
             
            segments.append([xseg, yseg, zseg]) 
            labels.append(class_label)

            
    return labels,segments
```

```python
labels, segments = get_segmentations(dfs[0])
```

```python
segments_array = np.array(segments, dtype=np.float64).reshape(-1,num_feature, window_size)
```

```python
print ("number of segments: " + str(len(labels))) 
print ("segements shape: " + str(segments_array.shape)) 
print ("labels for segments:" + str(list(set(labels)))) 
```

# Build model

```python
from keras import layers
from keras.layers import Input, Dense, Activation, ZeroPadding2D, BatchNormalization, Flatten, Conv2D
from keras.layers import AveragePooling2D, MaxPooling2D, Dropout, GlobalMaxPooling2D, GlobalAveragePooling2D
from keras.models import Model, load_model

from keras.models import Sequential
```

```python
def model():
    #create model
    model = Sequential()
    #add model layers
    model.add(Conv2D(64, kernel_size=2, activation='relu', input_shape=(3,520,1))) # 64 nodes, 3x3 filter matrix
    model.add(Conv2D(32, kernel_size=2, activation='relu'))
    model.add(Flatten())
    model.add(Dense(7, activation='softmax'))
    return model
```

```python
cnn_model = model()
cnn_model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
#train the model
cnn_model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=3)
```

```python
preds = cnn_model.evaluate(X_test, y_test)
print ("Loss = " + str(preds[0]))  
print ("Test Accuracy = " + str(preds[1]))
```

```python
def model2(input_shape):
    
    # Define the input placeholder as a tensor with shape input_shape. Think of this as your input image!
    X_input = Input(input_shape)

    # Zero-Padding: pads the border of X_input with zeroes
    X = X_input #ZeroPadding2D((2, 2))(X_input)

    # CONV -> BN -> RELU Block applied to X
    X = Conv2D(32, (2, 2), strides = (1, 1), name = 'conv0')(X)
    X = BatchNormalization(axis = 3, name = 'bn0')(X)
    X = Activation('relu')(X)

    # MAXPOOL
    X = MaxPooling2D((2, 2), name='max_pool')(X)

    # FLATTEN X (means convert it to a vector) + FULLYCONNECTED
    X = Flatten()(X)
    X = Dense(1, activation='softmax', name='fc')(X)

    # Create model. This creates your Keras model instance, you'll use this instance to train/test the model.
    model = Model(inputs = X_input, outputs = X, name='Model')

    return model
```

# Extract features

```python
class_labels = set(dfs[0]['label'].values.reshape(1,-1).squeeze())
class_labels.remove(0)
class_labels
```

```python
def extract_features(participant):
    
    """
    : participant -- label for the participants (0,1,2,..,14)
    : class_label -- different class (1,2,3,4,5,6,7)
    
    Return
    ======
    A new dataframe with the features [x_mean, x_var, x_mad, ..., z_mean, z_var, z_mad]
    """
    new_df = pd.DataFrame()
    
    # loop over class
    for class_label in class_labels: 
            
        df4class = dfs[participant][dfs[participant]['label'].values == class_label].head()
      
        means = []
        
        temp_df = pd.DataFrame()
        
        # loop over features
        for feature in features[1:4]: 
            temp_df[feature+'_mean'] = df4class[feature].mean().values 
            temp_df[feature+'_var'] = df4class[feature].var().values
        temp_df['label'] = class_label
            
        new_df = pd.concat([new_df,temp_df])
    
    
    return new_df
```

```python
df_concat = pd.DataFrame()
dfsNew = []
for p in range(0,15): 
    dfsNew.append(pd.concat([df_concat, extract_features(p)]))
    
```

```python
df_final = pd.concat([dfsNew[i] for i in range(0,15)], ignore_index=True)
dfclass = df_final.sort_values(by='label').reset_index().drop(['index'],axis=1)
```
